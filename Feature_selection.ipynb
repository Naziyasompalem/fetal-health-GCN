```python
from sklearn.model_selection import train_test_split
import pandas as pd

# List of features based on your description
selected_features = [
    'CLASS', 'LD', 'FS', 'DP', 'DP.1', 'ASTV', 'ALTV', 'SUSP', 'Variance', 'E', 'LB', 'DS.1',
    'DS', 'FM.1', 'DL', 'Min', 'DL.1', 'FM', 'Nzeros', 'Nmax', 'Max', 'Width', 'C', 'D', 'MSTV',
    'b', 'e', 'DE', 'Tendency', 'UC', 'UC.1', 'Median', 'AD', 'MLTV', 'Mean', 'A', 'Mode', 'B', 
    'AC', 'AC.1'
]  # All the features except the target 'NSP' and 'DR'

# Assuming 'data_second_sheet' is your DataFrame
X_selected = data_second_sheet[selected_features]
y = data_second_sheet['NSP']  # Target column

# Split dataset into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42, stratify=y)

# Check the shapes of the split data
print("Training data (X_train) shape:", X_train.shape)
print("Testing data (X_test) shape:", X_test.shape)
print("Training labels (y_train) shape:", y_train.shape)
print("Testing labels (y_test) shape:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, matthews_corrcoef
import time
import pandas as pd
```

```python
results = []

# Function to evaluate the models
def evaluate_model(model, model_name):
    start_time = time.time()  # Record the start time
    
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred, average='binary')  # For binary classification
    precision = precision_score(y_test, y_pred, average='binary')
    f1 = f1_score(y_test, y_pred, average='binary')
    mcc = matthews_corrcoef(y_test, y_pred)
    training_time = time.time() - start_time  # Time taken to train the model
    
    # Save results
    results.append({
        "Model": model_name,
        "Accuracy": accuracy,
        "Recall": recall,
        "Precision": precision,
        "F1 Score": f1,
        "MCC": mcc,
        "Time to Train": training_time
    })
```

```python
# Logistic Regression
log_reg = LogisticRegression(random_state=42)
evaluate_model(log_reg, "Logistic Regression")

# Support Vector Machine
svm = SVC(random_state=42)
evaluate_model(svm, "Support Vector Machine")

# Random Forest
rf = RandomForestClassifier(random_state=42)
evaluate_model(rf, "Random Forest")

# Decision Tree
dt = DecisionTreeClassifier(random_state=42)
evaluate_model(dt, "Decision Tree")

# Gradient Boosting
gb = GradientBoostingClassifier(random_state=42)
evaluate_model(gb, "Gradient Boosting")

# K-Nearest Neighbors
knn = KNeighborsClassifier()
evaluate_model(knn, "K-Nearest Neighbors")
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, matthews_corrcoef

# Function to evaluate the models
def evaluate_model(model, model_name):
    start_time = time.time()  # Record the start time
    
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Calculate metrics (use macro or weighted average for multiclass)
    accuracy = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred, average='macro')  # Changed to 'macro' for multiclass
    precision = precision_score(y_test, y_pred, average='macro')  # Changed to 'macro' for multiclass
    f1 = f1_score(y_test, y_pred, average='macro')  # Changed to 'macro' for multiclass
    mcc = matthews_corrcoef(y_test, y_pred)
    training_time = time.time() - start_time  # Time taken to train the model
    
    # Save results
    results.append({
        "Model": model_name,
        "Accuracy": accuracy,
        "Recall": recall,
        "Precision": precision,
        "F1 Score": f1,
        "MCC": mcc,
        "Time to Train": training_time
    })
```

```python
# Convert results to DataFrame for better presentation
results_df = pd.DataFrame(results)

# Display results
print(results_df)

# Optional: To visualize the results as a bar plot
import matplotlib.pyplot as plt

# Plot accuracy for comparison
fig, ax = plt.subplots(figsize=(10, 6))
results_df.set_index('Model').plot(kind='bar', ax=ax, stacked=True)

# Labeling and formatting
ax.set_title('Model Comparison')
ax.set_ylabel('Scores')
ax.set_xlabel('Models')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
